```python
import numpy as np
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,f1_score,recall_score,precision_score
```

```python
df_train=pd.read_csv('/content/fraudTrain.csv')
df_test=pd.read_csv('/content/fraudTest.csv')
```

```python
df=pd.concat([df_train,df_test])
```

```python
df_train.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['is_fraud'].value_counts()
```

```python
legit=df[df['is_fraud']==0]
fraud=df[df['is_fraud']==1]
```

```python
legit.shape
```

```python
fraud.shape
```

```python
legit_sample=legit.sample(n=9651)
```

```python
df_new = pd.concat([legit_sample, fraud], axis=0)
```

```python
df_new.head()
```

```python
df_new['is_fraud'].value_counts()
```

```python
unused_cols=['Unnamed: 0','first','last','unix_time','street','gender','job','dob','city','state','trans_num','merchant']
df_new.drop(columns=unused_cols,inplace=True)
```

```python
df_new.info()
```

```python
df_processed = pd.get_dummies(data=df_new)
df_processed
```

```python
X = df_processed.drop(columns='is_fraud', axis=1)
Y = df_processed['is_fraud']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y,random_state=42)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train,Y_train)
```

```python
y_pred1=lr.predict(X_test)
```

```python
print('Accuracy:',accuracy_score(Y_test, y_pred1))
print('Precision:',precision_score(Y_test, y_pred1))
print('Recall:', recall_score(Y_test, y_pred1))
print('F1 score:',f1_score(Y_test, y_pred1))
print('Confusion matrix:',confusion_matrix(Y_test, y_pred1))
```

```python
dt = DecisionTreeClassifier()
dt.fit(X_train, Y_train)
```

```python
y_pred2 = dt.predict(X_test)
```

```python
print('Accuracy:',accuracy_score(Y_test, y_pred2))
print('Precision:',precision_score(Y_test, y_pred2))
print('Recall:', recall_score(Y_test, y_pred2))
print('F1 score:',f1_score(Y_test, y_pred2))
print('Confusion matrix:',confusion_matrix(Y_test, y_pred2))
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, Y_train)
```

```python
y_pred3 = knn.predict(X_test)
```

```python
print('Accuracy:',accuracy_score(Y_test, y_pred3))
print('Precision:',precision_score(Y_test, y_pred3))
print('Recall:', recall_score(Y_test, y_pred3))
print('F1 score:',f1_score(Y_test, y_pred3))
print('Confusion matrix:',confusion_matrix(Y_test, y_pred3))
```

```python
RFC=RandomForestClassifier(random_state=0)
RFC.fit(X_train,Y_train)
y_pred4=RFC.predict(X_test)
```

```python
print('Accuracy:',accuracy_score(Y_test, y_pred4))
print('Precision:',precision_score(Y_test, y_pred4))
print('Recall:', recall_score(Y_test, y_pred4))
print('F1 score:',f1_score(Y_test, y_pred4))
print('Confusion matrix:',confusion_matrix(Y_test, y_pred4))
```

The best model which works here is Decision Tree Classifier.
